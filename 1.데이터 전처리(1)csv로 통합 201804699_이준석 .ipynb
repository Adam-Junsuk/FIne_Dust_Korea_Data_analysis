```python
preDataN = dust_df[['SO2', 'CO', 'O3', 'NO2', 'PM10']] #PM25는 결측치가 너무 많아 삭제해주었습니다.
```

```python
import numpy as np
from pandas import DataFrame, Series

preDataN=preDataN.replace({'SO2':-999}, {'SO2':np.nan})
preDataN=preDataN.replace({'CO':-999}, {'CO':np.nan})
preDataN=preDataN.replace({'O3':-999}, {'O3':np.nan})
preDataN=preDataN.replace({'NO2':-999}, {'NO2':np.nan})
preDataN=preDataN.replace({'PM10':-999.0}, {'PM10':np.nan})

#-999를 평균값으로 대체하기 위해 NaN으로 바꿔줬습니다.
```
